In questo notebook definisco le funzioni necessarie a:
    1. Estrarre singoli dataframe da quelli aggregati
    2. salvare tutto in csv
    

```python
import pandas as pd
import numpy as np
import glob
import re
```

```python
#questa funzione pulisce il df eliminando le colonne inutili e droppa i NaN

def get_clean_dataframe(path):
  df = pd.read_csv(path, low_memory= False)
  if 'Unnamed: 7' in df.columns:
    df = df.drop(columns=['Unnamed: 7'])
  df = df.drop(columns=['Round','View', 'Unnamed: 0', 'Name', 'Name.1', 'ELO', 'Points'])
  df = df.dropna()
  return df
```

```python
#questa funzione prende in input un df in cui sono aggregati più anni
# e ne estrae i singoli df
def separa_anni(df, year_list):
  df['Date'] = pd.to_datetime(df['Date'], format='mixed', errors='coerce')
  dataframe_list = []
  # Estrai DataFrames per ogni anno
  for year in year_list:
    '''if (year == year_list[-1]):
      break'''

    df_year = df[df['Date'].dt.year == year]
    dataframe_list.append(df_year)
  return(dataframe_list)

```

```python
def salva_df_singoli (lista, lista_anni):
  for i in range(len(lista)):
    for anno in lista_anni:
      lista[i].to_csv(f'./chess_data_{anno}.csv')
```

```python
dir = '/content/'
df_tot = pd.DataFrame()
to_append = []
for file in glob.glob('./*.csv'):
  path=f'{dir}/{file}'
  df_to_append = get_clean_dataframe(path)
  to_append.append(df_to_append)
  df_tot = pd.concat (to_append)
```

```python
df_tot.to_csv('./chess_data_9598.csv')
```

```python
#df_aggregati = get_clean_dataframe(path)
lista_df = separa_anni(df_aggregati, anni)
salva_df_singoli(lista_df, anni)
```
