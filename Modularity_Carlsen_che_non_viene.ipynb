```python
def Community_making(G):
    """Restituisce i giocatori (chiavi) con le rispettive community di appartenenza (valore)"""
    partition = community_louvain.best_partition(G, random_state=42)
    return partition

def getPartitionwithColpe(G):
    carlsen = 'Colpe, M'
    partition_w_carlsen = {}
    communities= Community_making(G)
    #print(communities)
    players_same_comm=[]
    complement=[]
    for key, value in communities.items():
        if value == communities[carlsen]:
            players_same_comm.append(key)
        else:
            complement.append(key)
    return [players_same_comm+complement]

def getPartitionwithCarlsen(G):
    carlsen = 'Carlsen, M'
    partition_w_carlsen = {}
    communities= Community_making(G)
    #print(communities)
    players_same_comm=[]
    complement=[]
    for key, value in communities.items():
        if value == communities[carlsen]:
            if key!='Brault, G':
                players_same_comm.append(key)
        else:
            if key!='Brault, G':
                complement.append(key)
    return [players_same_comm+complement]

def central_cluster_of_network(G):
    largest_cc = max(nx.connected_components(G), key=len)
    cc = G.subgraph(largest_cc).copy()
    return cc
```
