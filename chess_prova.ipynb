```python
def import_many_data(list_of_data, is_what, num_table):
    
    """
        is_what : write the file format (example csv, html...) you want to read
        num_table : write the number of corresponding table you want to print out from your dataset
    """
    
    # Creating empty lists to save "pd.read" and 
    # to make dataframes in that list using pd.DataFrame
    
    data = []
    df_list = []
    
    if is_what == "html":
        for i in range(len(list_of_data)):
        
            # Appending elements
        
            data.append(pd.read_html(list_of_data[i]))
            df_list.append(pd.DataFrame(data[i][num_table]))
        else:
            if is_what == "csv":
                for j in range(len(list_of_data)):
        
                    # Appending elements
        
                    data.append(pd.read_csv(list_of_data[j]))
                    df_list.append(pd.DataFrame(data[j][num_table]))
                
    df_list = pd.concat(df_list)
    return df_list

tournament_1 = [for i in ... 'https://www.365chess.com/tournaments/World_Blitz_2019_2019/43856' + str()]
list_of_data = ['https://www.365chess.com/tournaments/World_Blitz_2019_2019/43856',
               'https://www.365chess.com/tournaments/30th_Cracovia_Open_A_2019/43861',
                'https://www.365chess.com/tournaments/World_Rapid_2019_2019/43853']

import_many_data(list_of_data, "html", 3)
```
