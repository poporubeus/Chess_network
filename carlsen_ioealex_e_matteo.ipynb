```python
def get_white(df):
    wcarlsen = []
    for i in range(len(df['White'])):
        if df['White'][i]=='Carlsen, M':
            wcarlsen.append(df['Black'][i])
    return wcarlsen

def get_black(df):
    bcarlsen = []
    for i in range(len(df['Black'])):
        if df['Black'][i]=='Carlsen, M':
            bcarlsen.append(df['White'][i])
    return bcarlsen
```

```python
def conta_avversari_carlsen(df):
    wcarlsen = get_white(df)
    bcarlsen = get_black(df)
    total_avversari = wcarlsen + bcarlsen
    return list(set(total_avversari))
```

```python
lista_avversari=[]
```

```python
x = np.concatenate((np.arange(2005,2009,1), np.arange(2010,2023,1)))

avversari_array = np.array([len(lista_avversari[i]) for i in range(len(lista_avversari))])
plt.plot(x, avversari_array, marker='o', color='blue')
plt.xticks(np.round(np.concatenate((x,np.array([2009]))),0),rotation='vertical')
plt.xlabel('Year')
plt.ylabel('Number of opponents by year')
plt.show()
```

```python
def conta_partite(dataframe):
    matchesw = dataframe[dataframe['White']=='Carlsen, M'].count()
    matchesb = dataframe[dataframe['Black']=='Carlsen, M'].count()
    return matchesw + matchesb
```

```python
partite_totali = []
for i in range(5,23):
    if i ==9:
        pass
    else:
        partite_totali.append(conta_partite(globals()['df_%s'%i])[0])
```

```python
partite_totali = np.array(partite_totali)
```

```python
plt.plot(x, partite_totali, marker='o', color='orange', label='Number of games by year')
plt.plot(x, avversari_array, marker='P', color='blue', label='Number of opponents by year')
plt.xticks(np.round(np.concatenate((x,np.array([2009]))),0),rotation='vertical')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Values')
plt.show()
```

```python
normalized_arr = partite_totali / avversari_array
```

```python
plt.plot(x, normalized_arr, marker='o', color='red')
plt.xticks(np.round(np.concatenate((x,np.array([2009]))),0),rotation='vertical')
plt.xlabel('Year')
plt.ylabel('Games-Opponents ratio')
plt.show()
```

```python
count = np.zeros(len(lista_avversari)-1)
var = True
tot_count = np.zeros(len(lista_avversari)-1)
for i in range(1,len(lista_avversari)):
    for name in lista_avversari[i]:
        for j in range(i):
            if name in lista_avversari[j]:
                var = False
        if var == True:
            count[i-1] = count[i-1]+1
        var = True
    tot_count[i-1] = np.sum(count)
```
